#**조건문**

* 조건에 따라 다른 동작을 수행하게 함

* 예) 시험 점수에 따라 출력문장을 다르게 하는 경우:

> <img src="https://drive.google.com/uc?id=1zkGrj1f-OHEtMxskxJotQKIlvda1jlIx" width="400"/>

* 위 예의 코드는 아래와 같음:

  * **실행 문장은 들여쓰기해야 함** 

  * 들여쓰기 레벨이 같은 연속된 문장들을 **블록 (block)**이라 함 

```
score = 60

if (score >= 60) : 
  print("합격입니다.")
```

```
if (score >= 60) : print("합격입니다.")
```

```
score = 40

if (score >= 60) :
  print("합격입니다.")
else :
  print("불합격입니다.")
```

```
score = 90

if (score >= 90) :
  print("합격입니다.")  
  print("장학금도 받을 수 있습니다.")
elif (score >= 60) :
  print("합격입니다.")
else :
  print("불합격입니다.")
```

* 블록 안에 새로운 if 문을 만드는 것이 가능

```
score = 90

if (score >= 60) :
  print("합격입니다.")

  if (score >= 90) : 
    print("장학금도 받을 수 있습니다.")
  else :
    print("장학금은 받을 수 없습니다.")
      
else :
  print("불합격입니다.")
```

#**반복문**

* 문장(또는 블록)을 원하는 회수 만큼/값 범위 안에서 반복

```
# range(시작값,종료값,변화량) 함수를 이용해 반복시 x가 갖게될 값을 설정

for x in range(5) :
  print(x)
  print("환영합니다.")
```

```
list(range(1,5,2))
```

```
for x in range(5) :
  print(x)
```

```
for x in range(0,5) :
  print(x)
```

```
for x in range(1,5,2) :
  print(x)
```

```
# 값의 영역을 리스트로 직접 줄 수 있음

for name in ["철수", "영희", "길동", "유신" ] :   
	 print("안녕! " + name)
```

* 문장(또는 블록)을 특정 조건이 만족하는 동안 반복

```
i = 0
while (i < 5) : 
      print("환영합니다.")
      i = i + 1
```

* 블록 안에 새로운 반복 문을 만드는 것이 가능

```
for y in range(5) :
	for x in range(10) :
		print("*", end="")
	print("")	
```
