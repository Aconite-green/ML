# 함수 (function)

* 특정 기능을 하는 문장 블록을 따로 떼서 이름을 붙인 것

  * 자주 쓰는 기능이라면 매번 블록을 타이핑할 필요 없이 함수를 호출하면 끝

* 함수 이름을 정하는 규칙은 변수의 경우와 동일



```
def poem_flower():
  print("자세히 보아야 예쁘다")
  print("오래 보아야 사랑스럽다")
  print("너도 그렇다")
```

```
poem_flower()
poem_flower()
```

* 매개변수를 정의하여 함수의 몸체 블록에 데이터를 전달할 수 있다.

```
def say_hello(name, msg):
		print('안녕, ', name, '야, ', msg, '하자')
```

```
say_hello('철수', '게임')
say_hello('영희', '공부')
```

* 매개변수의 디폴트(default) 값을 정의할 수 있음

```
def say_hello(name, msg="게임"):
		print('안녕, ', name, '야, ', msg, '하자')
```

```
say_hello("철수")
say_hello('영희', '공부')
```

* 함수 실행 후 계산한 값을 반환하게 할 수 있음

```
# 주어진 시작값부터 끝값까지의 모든 정수를 더한 값을 리턴하는 함수

def get_sum(start, end) :
	sum = 0
	for i in range(start, end+1) :
		sum += i
	return sum
```

```
value = get_sum(1, 10)

print(value)
```

* 복수 개의 값을 리턴하는 것이 가능

```
# 주어진 매개변수값의 제곱, 세제곱, 네제곱 값을 리턴하는 함수

def get_pow(x) :
  y2 = x**2
  y3 = x**3
  y4 = x**4
  
  return (y2, y3, y4)
```

```
z2, z3, z4 = get_pow(3)

z2, z3, z4
```

# 수학 함수의 사용

* math 모듈을 임포트해서 필요한 수학 함수를 사용할 수 있음

* API (Application Programming Interface) 정보 찾기 : http://overapi.com/python

```
# math 모듈의 수학 함수 임포트

from math import sin, cos, pi
```

```
sin(1)
```

```
cos(2*pi)
```

```
# x 값을 -2부터 2까지 0.01 간격으로 생성하여 리스트 객체에 담기

x = [-2]

for i in range(-199,201) :

  x += [0.01 * i] 
```

```
x
```

```
# 축약된 버전

x = [0.01*i for i in range(-200,201)]
```

```
# -2pi 부터 2pi까지에 대해 sin 함수 값 구하기

y_sin = [sin(j*pi) for j in x]

y_sin
```

# 그래프 그리기

* matplotlib 모듈 임포트해서 사용

```
import matplotlib.pyplot as plt

plt.plot(x,y_sin)
plt.show()
```

# 그래프의 스타일 설정하기

* matplotlib 모듈의 rcParams를 설정

```
import matplotlib.pyplot as plt

plt.rcParams['figure.figsize'] = (6,6)
plt.rcParams['lines.linewidth'] = 10

fig = plt.plot(x,y_sin)

plt.xlabel('xlabel')
plt.ylabel('ylabel')

plt.show()
```
