#**변수**

* 변수는 데이터를 저장할 수 있는 **메모리의 기억공간**이다.

* 변수에 값을 저장하는 순간부터 변수(메모리 공간)가 생성됨

 * C언어에서는 명시적으로 변수를 선언하고 생성해야 함
 
* 변수에 값을 저장할 때는 **대입연산자(=)**를 사용한다.

 * 대입연산자는 오른쪽의 값을 왼쪽의 변수에 저장한다.

 * 대입연산자 왼쪽에는 값을 저장할 기억공간(변수)이 온다.

```
x
```

```
x = 10
```

```
x
```

```
print(x)
```

```
print(X)
```

```
x = 100

y = 3.14

x * y
```

 * 대입연산자 오른쪽에 변수가 올 때 : 변수의 값을 왼쪽의 변수에 복사한다.


```
y = x

print(y)
```

```
y
```

```
10 = 20
```

```
width = 10

height = 20

area = width * height

print(area)
```

# 변수의 이름 정하기

* 의미 있는 이름을 사용
* 소문자와 대문자는 서로 다르게 취급된다.
* 변수의 이름은 영문자와 숫자, 밑줄(_)로 이루어진다.
* 변수 이름의 첫글자로 숫자가 올 수 없다. 
* 변수의 이름 중간에 공백이 들어가면 안된다. 
* 단어를 구분하려면 밑줄(_)을 사용하면 좋다.

# 딕셔너리 (Dictionary)

* 키(key)와 값(value)의 쌍을 저장할 수 있는 객체
* 사전에 단어(key)와 의미(value)가 쌍으로 기록되어 있는 것과 동일

> <img src="https://drive.google.com/uc?id=1siNyw7uexToSekhCzmY1eLMnrm2l1Hrw" width="300"/>


```
contacts = {'Kim':'01012345678',  'Park':'01012345679', 'Lee':'01012345680' }

print(contacts)
```

```
# 특정 키의 값 가져오기

contacts['yun']
```

```
contacts.get('yun')
```

```
# 기존 항목 수정

contacts['Kim'] = '01056781234'

contacts['Kim']
```

```
# 항목 추가

contacts['Choi'] = '01056781234'

contacts
```

```
# 항목 삭제

contacts.pop('Kim')

print(contacts)
```
